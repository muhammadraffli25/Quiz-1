### Membuat Data Frame

```python
# Dataframe dari dictionary
data = {
    "Name" : ["Alice", "Bob", "Charlie"],
    "Age"  : [25,30,25],
    "City" : ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Dataframe dari list
data_list = [
    ["David", 28, "Miami"],
    ["Eva", 22, "Seattle"],
     ["frank", 33, "Boston"]
]

df = pd.DataFrame(data_list, columns=["Name", "Age", "City"])
print(df)
```

```python
print(df.columns.to_list())
```

```python
df.columns = ["nama", "umur", "kota"]
print(df)
```

```python
print(df["nama"])
```

```python
print(df[["nama","kota"]])
```
